어떤 확률 변수 $Z$가 있고 $X=g(Z)$관계에 의해 정의되는 확률변수 $X$가 있다. 
이때 확률 변수 $Z$가 $z_{1} \le z \le z_{2}$의 값을 가질 때 발생하는 사건을 A라 하자. 예를 들어  $z_{1} \le z \le z_{2}$이면 우리는 오늘 저녁에 피자를 먹는다. 즉, 피자를 먹는 사건이 A이다. 그렇다면 오늘 저녁 피자를 먹을 수 있을 확률은 다음과 같다.

$$P(A) = P(z_{1} \le z \le z_{2}) =\int _{z_{1}} ^{z_{2}} p_{Z}(z) dz \tag{1}$$

이제 같은 사건 A가 일어날 확률을 $X=g(Z)$로 변환된 확률변수 $X$로 표현하면 다음과 같을 것이다.

$$P(A) = P(x_{1} \le x \le x_{2}) = \int _{g(z_{1})} ^{g(z_{2})} p_{X}(x) dx \tag{2}$$

아직 변환된 $X$에 대한 확률분포 $p_{X}(x)$를 모르는데 두 경우 사건 A가 일어날 확률을 동일하므로 $X=g(Z)$관계에 의해 적당히 잘 변환된 $p_X(x)$를 다음과 같은 관계식으로부터 찾을 수 있다.<sup>[1]</sup>

$$
P(z_{1} \le z \le z_{2}) = \int _{z_{1}} ^{z_{2}} p_{Z}(z) dz = \int _{x_{1}=g(z_{1})} ^{x_{2}=g(z_{2})} p_{X}(x) dx = P(x_{1} \le x \le x_{2}) 
\tag{3}
$$

일단 $z_{1} \le z \le z_{2}$ 전구간에서 $X=g(Z)$의 역함수 $Z=g^{-1}(X)$가 존재한다고 가정하면 다음처럼 변수를 바꿔서 적분할 수 있다.

$$
\int _{z_{1}} ^{z_{2}} p_{Z}(z) dz = \int _{x_{1}=g(z_{1})} ^{x_{2}=g(z_{2})} p_{Z}(g^{-1}(x)) \frac{d g^{-1}(x)}{dx} dx \tag{4}
$$

확률변수 $Z$와 $X$의 도메인이 모두 1차원이라면 식(4)와 같으나 2차원 이상이 되어 적분의 경계가 영역이 되면 다음처럼 Jacobian에 절대값을 적용해야한다. 확률변수 $Z$와 $X$의 특정 영역을 $\hat{Z}$, $\hat{X}$라 하면

$$
\int _{z \in \hat{Z}} p_{Z}(z) dz = \int _{x \in \hat{X}} p_{Z}(g^{-1}(x)) \left| \frac{d g^{-1}(x)}{dx} \right| dx \tag{5}
$$

1차원에서 절대값이 필요없는 이유는 적분구간 양끝의 순서를 바꿀 수 있기 때문이다.
예를 들어

$$
\int_{1}^{2} (3x+1)^3 dx = \frac{715}{4}
$$

에서 $x = g(t)= -\frac{t+1}{3}$으로 치환한다고 하면 $t = -3x-1$ 이므로 

$$
\int_{1}^{2} (3x+1)^3 dx=\int_{-3(1)-1}^{-3(2)-1} \left(3g(t) +1\right)^3 \frac{dg(t)}{dt} dt = \int_{-4}^{-7} (-t)^3 \left(-\frac{1}{3}\right) dt = \frac{715}{4}
$$

가 되는데 변수를 바꾸고 나서 적분의 경계를 보면 큰수가 아래오고 작은 수가 위로 가 있다. 이는 $g(t)$를 감소함수로 선택했기 때문에 일어나는 현상인데 이때 적분의 경계의 위치를 바꾸어 작은수가 밑으로 오고 큰수가 위로 가게 하면

$$
\int_{-4}^{-7} (-t)^3 \left(-\frac{1}{3}\right) dt = -\int_{-7}^{-4} (-t)^3 \left(-\frac{1}{3}\right) dt = \int_{-7}^{-4} (-t)^3 \left(\frac{1}{3}\right) dt=\frac{715}{4}
$$

가 되어 경계를 바꾸는것으로 인해 $dg(t) / dt$의 음의 부호가 사라진다. 그래서 항상 작은 수를 아래쪽 경계로 쓰기로 하고, $g(t)$가 증가함수든 감소함수든 상관이 없으려면 식(5)와 같이 절대값을 써준다. 하지만 1차원에서는 경계의 아래쪽이 크든 작든 상관없으므로 $g(t)$가 어떤 함수라하더라도 절대값을 안써도 상관없다. 

어쨋거나 식(5)의 우변은 식(3)에 의해 아래와 같으므로

$$
\int _{x \in \hat{X}} p_{Z}(g^{-1}(x)) \left| \frac{d g^{-1}(x)}{dx} \right| dx 
= \int _{x \in \hat{X}} p_{X}(x) dx   \tag{6}
$$

결과적으로 $p_X(x)$는 다음과 같다.

$$p_X(x) = p_Z (g^{-1}(x)) \left|\frac{d g^{-1}(x) }{d x}\right| \tag{7}$$

$p_X(x)$를 교재에 따라 다음처럼 다르게 표현하기도 하는데 다 같은 표현이다. <sup>[2][3]</sup>

$$
p_X(x) 
= p_Z (g^{-1}(x)) \left|\frac{d g^{-1}(x) }{d x}\right| 
= p_Z (g^{-1}(x)) \left|\frac{dz }{d g(z)}\right| 
= \frac{p_Z (g^{-1}(x))}{\left|\frac{d g(z)}{dz}\right|}
= \frac{p_Z (z)}{\left|\frac{d g(z)}{dz}\right|} 
\tag{8}
$$

반대로 $p_X(x)$로 부터 $p_Z(z)$는 다음과 같이 구해진다.

$$
p_Z(z) = p_X(x)\left|\frac{dg(z)}{dz}\right|=p_X(g(z))\left|\frac{d g(z)}{dz}\right| \tag{9}
$$

여기서 알 수 있는 점은 $p_X(x)$가 $Z=g^{-1}(X)$ 관계에 의해 단순히 합성함수인 $p_Z(g^{-1}(x))$ 와 같지 않다는 점이다. $X=g(Z)$이니까  $p_X(x)$는  $p_Z(z)$에서 $z$를 모두 $g^{-1}(x)$로 대체한 합성함수  $p_Z(g^{-1}(x))$가 아닐까 생각할 수 도 있는데 아니라는 것이다. 마찬가지로 $p_Z(z)$도 $X=g(Z)$ 관계에 의한 합성함수 $p_X(g(z))$ 와 같지 않다. 실제 코딩 실습으로 알아보면 다음과 같다. 아래 실습예제는 PRML<sup>[4]</sup>의 연습문제 1.4번을 살짝 거꾸로 바꾼것이다.

확률변수 $Z$의 확률밀도함수는 평균이 6, 분산이 1.0인 정규 분포이다. 

$$
p_{Z}(z) = \mathcal{N}(z \, | \, \mu = 6, \sigma^{2}=1.0) \tag{10}
$$

$Z$를 $X$로 매핑하는 매핑함수 $g(z)$는 다음과 같다. 흔히 쓰는 시그모이드 함수이다.

$$
x = g(z)= \frac{1}{1+\exp (-z +5) } \tag{11}
$$

역함수 $g^{-1}(x)$는 다음과 같다.

$$
z = g^{-1}(x)= \log(x)-\log(1-x)+5 \tag{12}
$$

이를 바탕으로 식(7)에 의해 $X$의 pdf를 구할 수 있다.
다음에 실습코드가 있고 class pz_gen으로 연속확률변수 $Z$를 class px_gen로 확률변수 $X$를 구현했다.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.stats import rv_continuous
from matplotlib.patches import Polygon

def g(z) :
    return 1.0 / (1.0+np.exp(-z+5))

def g_inv(x) :    
    return np.log(x)-np.log(1-x)+5

class pz_gen(rv_continuous):
    "P_Z(z) distribution for R.V. Z"
    
    mu = 6
    sigma = 1.0
    
    def Normal(self, z) :
        return (1 / np.sqrt(2*np.pi*(self.sigma**2)))*np.exp( -(1/2*(self.sigma**2))*(z-self.mu)**2 )

    def _pdf(self, z) :
        return self.Normal(z)

        
class px_gen(rv_continuous):
    "P_X(x)  distribution for R.V. X"

    mu = 6
    sigma = 1.0
    
    def __init__(self):
        """
        확률 변수 X는 범위가 0~1사이에 있다.
        """
        rv_continuous.__init__(self, a=0.0, b=1.0)
        
    def Normal(self, z) :
        return (1 / np.sqrt(2*np.pi*(self.sigma**2)))*np.exp( -(1/2*(self.sigma**2))*(z-self.mu)**2 )
    
    def g(self, z) :
        return 1.0 / (1.0+np.exp(-z+5))

    def g_prime(self, z):
        return self.g(z)*(1-self.g(z))
    
    def g_inv(self, x) :    
        return np.log(x)-np.log(1-x)+5

    def g_inv_prime(self, x):
        return (1/x)+(1/(1-x))
        
    def _pdf(self, x) :
        #아래 두 표현 모두 동일하다. EQ.(8)
        #return np.multiply( self.Normal(self.g_inv(x)) , np.abs(1/self.g_prime(self.g_inv(x))) )
        return np.multiply( self.Normal(self.g_inv(x)) , np.abs(self.g_inv_prime(x)) )
    
# Get current size
fig_size = plt.rcParams["figure.figsize"]
fig_size = (15,10)
plt.rcParams["figure.figsize"] = fig_size
font = { 'size': 20, }

#각 확률변수의 샘플 개수
nos = 5000

Z = pz_gen()
X = px_gen()
z = np.linspace(0, 10, 100)
x = g(z)
z_hat = 6
x_hat = g(z_hat)

Z_sample = Z.rvs(size=nos)
X_sample = X.rvs(size=nos)

fig, ax2 = plt.subplots(1, 1)

ax2.plot(z, Z._pdf(z), 'r')
ax2.plot(z, x)
ax2.plot(X._pdf(x), x, 'm')
ax2.text(7.5, 0.85, r'$x=g(z)$', color='orange', fontdict=font)
ax2.text(1.8, 0.65, r'$p_X(x)$', color='m', fontdict=font)
ax2.text(2.0, 0.4,  r'$p_Z(g^{-1}(x))$', color='green', fontdict=font)
ax2.text(7.3, 0.2,  r'$p_Z(z)$', color='r', fontdict=font)
#극점 잇는 점선 그리기
ax2.plot( (z_hat, z_hat) , (Z.Normal(z_hat),x_hat), 'y--' )
ax2.plot( (4 , z_hat),     (  x_hat , x_hat  ), 'y--')
#극점 그리기
ax2.plot(z_hat, Z.Normal(z_hat), 'ro')
ax2.plot(z_hat, x_hat, 'ro')
ax2.hist(Z_sample, 40, normed=1, facecolor='r', alpha=0.5, edgecolor='r')
ax2.hist(X_sample, 40, normed=1, facecolor='m', alpha=0.5, edgecolor='m', orientation='horizontal')
ax2.set_xlim([0, 10])
ax2.set_xlabel('z', fontsize=20)

# Make the shaded integral region P_z(z)
ix = np.linspace(5.5, 6.5)
iy = Z._pdf(ix)
verts = [(5.5, 0)] + list(zip(ix, iy)) + [(6.5, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.6', alpha=0.5)
ax2.add_patch(poly)

# Make the shaded integral region P_x(x)
ix = np.linspace(g(5.5), g(6.5))
iy = X._pdf(ix)
verts = [(0,g(5.5))] + list(zip(iy, ix)) + [(0,g(6.5))]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.6', alpha=0.5)
ax2.add_patch(poly)

ax3 = ax2.twiny()
ax3.plot(Z.Normal(g_inv(x)),   x, 'g') 
ax3.plot(X.Normal(g_inv(x_hat)), x_hat, 'ro') #극점 그리기
ax3.set_xlim([0, 1])
ax2.set_ylabel('x', fontsize=20)

plt.show()

#print("dx[p_x(x)]=0 at x=6 , dz[p_x(g(z))]=0 at g_inv(6)={:f}".format(z_hat))

#구간 적분을 해본다.
print("우리가 저녁으로 피자를 먹을 확률은")
print("integral from 5.5     to 6.5    p_Z(z) dz = {:f}".format(  integrate.quad(Z._pdf, 5.5, 6.5)[0]   ) )
print("integral from g(5.5)  to g(6.5) p_X(x) dx = {:f}".format(  integrate.quad(X._pdf, g(5.5), g(6.5))[0]   ) )
print("동일하다.")
```

결과로 나온 그래프는 가로축을 $z$로 하여 $p_{Z}(z)$와 $X=g(Z)$를 그리고 세로축에 매핑된 $X$에 대한 확률밀도함수를 그린것이다. $Z$가 $X=g(Z)$의해 변환될 때 $p_{X}(x)$는 그림에서 마젠타색 그래프가 된다. $p_{Z}(z)$에서 $X=g(Z)$를 이용하여 만든 합성함수 $p_{Z}(g^{-1}(x))$는 녹색으로 표시되었는데 실제 $p_{X}(x)$와는 다름을 확인 할 수 있다. $p_{Z}(z)$의 극점은 합성함수 $p_{Z}(g^{-1}(x))$의 극점과 일치하나 실제 $p_{X}(x)$의 극점과는 일치하지 않음을 알 수 있다. 마지막으로 음영처리된 부분이 우리가 저녁으로 피자를 먹을 확률에 해당하는 부분인데 두 확률분포에서 음영처리된 부분의 면적은 같으므로 변환된 확률변수상에서 사건 A가 일어날 확률은 동일하다는 전재가 깨어지지 않았음을 확인할 수 있다.

머신러닝<sup>[5]</sup>책에 나오는 간단한 예제 하나를 더 정리해 본다.

$X \sim Unif(-1,1)$인 확률변수에 대해 $Y = X^2$인 관계에 있는 확률변수 $Y$의 확률밀도함수를 구하라.

변환 관계는 다음과 같다.

$$
y = f(x) = x^2 \\[5pt]
x = f^-1 (y) = \sqrt{y}
$$

그리고 논의한바와 같이 확률밀도함수의 변환은 다음과 같다.

$$
p_y(y) = p_x(x) \left| \frac{dx}{dy} \right| = p_x(f^{-1}(y)) \left| \frac{df^{-1}(y)}{dy} \right|
$$


- 균등분포의 확률밀도함수의 정의는 $\frac{1}{b-a}$이므로 확률변수 $X$의 확률밀도함수는 

$$
p_x(x) = \frac{1}{b-a} = \frac{1}{1-(-1)} = \frac{1}{2}
$$

- 확률변수 $Y$의 도메인은 $Y = X^2$에 의해 (0,1)이다.

- $p_x(f^{-1}(y))$는 $p_x(x)$의 $x$를 $x=f^{-1}(y)$에 의해 $y$로 바꾼 합성함수이다. 따라서 변환된 도메인 (0,1)에 의해 


$$
p_x(f^{-1}(y)) = \frac{1}{b-a} = \frac{1}{1-0} = 1
$$

이고, 최종적으로 다음과 같이 $Y$의 확률밀도함수를 구할 수 있다.

$$
p_y(y) = p_x(f^{-1}(y)) \left| \frac{df^{-1}(y)}{dy} \right| = 1 \frac{d \sqrt{y}}{dy} = \frac{1}{2} y^{\frac{1}{2}}
$$
